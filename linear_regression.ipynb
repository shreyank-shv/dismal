# Step 2: Data Preprocessing

* Drop Loud Cover and Daily Summary

```python
df.drop(['Loud Cover', 'Daily Summary'], axis=1)
```

## Formatted Date currently has object dtype, convert into datetime format

```python
import datetime

df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['Formatted Date'].sample(5)
```

```python
df['year'] = df['Formatted Date'].dt.year
df['month'] = df['Formatted Date'].dt.month
df['day'] = df['Formatted Date'].dt.day
df['weekday'] = df['Formatted Date'].dt.weekday
df.head(5)
```

```python
df = df.drop(['Loud Cover', 'Daily Summary'], axis=1)
df.head(5)
```

# Step-3: Create Pipeline for Category Encoding and Model Predictions

```python
df['Precip Type']= df['Precip Type'].dropna()
df['Precip Type'].unique()
```

```python
df['Precip Type'] = df['Precip Type'].fillna("None")
df["Precip Type"] = df['Precip Type'].astype('str')
df['Summary'] = df['Summary'].astype('str')
df["Precip Type"].dtype, df["Summary"].dtype
```

```python
from sklearn.preprocessing import OrdinalEncoder
from category_encoders.james_stein import JamesSteinEncoder
```

```python
ordinal_columns = ['Precip Type', 'Summary']
#jstein_columns = ['Summary']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(), ordinal_columns),
        #('cat', JamesSteinEncoder(), jstein_columns)
    ])

model = LinearRegression()

pro_pipeline = Pipeline([('preprocessor', preprocessor),('model', model)])
```

```python
X = df.drop(['Apparent Temperature (C)'], axis=1)
X= X.drop(['Formatted Date'], axis=1)
y = df['Apparent Temperature (C)']
X.head(5)
```

```python
cat_encoder = LabelEncoder()
ord_encoder = OrdinalEncoder()

X['enc_summary'] = cat_encoder.fit_transform(X['Summary'])
X['ord_summary'] = ord_encoder.fit_transform(X['Summary'].values.reshape(-1,1))
X['ord_precip'] = ord_encoder.fit_transform(X['Precip Type'].values.reshape(-1,1))
X = X.drop(['Summary', 'Precip Type', 'enc_summary'], axis=1)
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
print(score)
```
